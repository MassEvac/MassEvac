```python
settings = '30min_30cm'
accessibility = pandas.read_csv('flood/OA_accessible_{}.csv'.format(settings),header=None,index_col=0).T
accessibility.index = range(10000)
accessibility
```

```python
data = pandas.read_csv('../v4/flood/inflows.csv',header=0,index_col=None)
inflows = data[['Eden','Petterill','Caldew']]
inflows
```

```python
def subhist(key,color):
    key_inflows = inflows[key].values
    bins=np.logspace(np.log10(min(key_inflows)),  np.log10(max(key_inflows)), 50)
    plt.hist(key_inflows,bins=bins,histtype='step',color=color,label=key,normed=True,log=False)
    plt.xscale('log')
    plt.legend(fontsize=12)
    plt.xlabel('$\mathrm{m^3/s}$',fontsize=20)
    print key,np.mean(key_inflows),np.std(key_inflows)

plt.figure(figsize=(15,3))

plt.subplot(131)
subhist('Eden','r')

plt.subplot(132)
subhist('Caldew','g')

plt.subplot(133)
subhist('Petterill','b')

plt.savefig('flood/figs/inflows-subplots.pdf', bbox_inches='tight' )
```

```python
import random

try:
    train = pandas.DataFrame.from_csv('flood/train_indices.csv').values.flatten()
    test = pandas.DataFrame.from_csv('flood/test_indices.csv').values.flatten()
except IOError:
    train = random.sample(range(10000),5000)
    test = [i for i in range(10000) if i not in train]
    pandas.DataFrame(train).to_csv('flood/train_indices.csv')
    pandas.DataFrame(test).to_csv('flood/test_indices.csv')

train_inflows = data.ix[train]
test_inflows = data.ix[test]

train_inflows.to_csv('flood/train_inflows.csv')
test_inflows.to_csv('flood/test_inflows.csv')
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from scipy import stats as ss

train_target = {}
test_target = {}

target_mean = {}
target_std = {}

for i in accessibility.columns:    
    target = (accessibility[i]*10).round()
    train_target[i] = np.array(target.ix[train],dtype=int)
    test_target[i] = np.array(target.ix[test],dtype=int)
    target_mean[i] = target.mean()
    target_std[i] = target.std()
    print i,target_mean[i],target_std[i]

with open('flood/target_mean_{}.csv'.format(settings),'w') as file:
    writer = csv.writer(file, delimiter=',')
    for key,value in target_mean.iteritems():
        writer.writerow([key,value])    
        
with open('flood/target_std_{}.csv'.format(settings),'w') as file:
    writer = csv.writer(file, delimiter=',')
    for key,value in target_std.iteritems():
        writer.writerow([key,value])
```

```python
y_pred = {}
pc_correct = {}
r_squared = {}
    
for i in accessibility.columns:    
    gnb = GaussianNB()
    y_pred[i] = gnb.fit(train_inflows, train_target[i]).predict(test_inflows)
    pc_correct[i] = (test_target[i] == y_pred[i]).sum()/float(len(y_pred[i]))*100
    r_squared[i] = ss.pearsonr(test_target[i],y_pred[i])[0]**2    
    
    print i,pc_correct[i],r_squared[i]

import csv
with open('flood/pc_correct_{}.csv'.format(settings),'w') as file:
    writer = csv.writer(file, delimiter=',')
    for key,value in pc_correct.iteritems():
        writer.writerow([key,value])
```
