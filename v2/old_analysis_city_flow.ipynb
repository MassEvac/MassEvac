```python
# Load a list of cities and their e
import cities
import pickle
import scipy.stats as ss
from collections import OrderedDict

sim_name = 'massevac'
all = cities.all

scenario = 'ia'
removed_ia = OrderedDict()
width = OrderedDict()
DX90 = OrderedDict()
n = OrderedDict()

# Load files created by 
for city in all:

    fname = 'analysis/anders/{0}/{1}'.format(scenario,city)
    
    file = open(fname, 'r')
    removed_ia[city],width[city],DX90[city],n[city] = pickle.load(file)
    file.close()
```

```python
scenario = 'ff'

removed_ff = OrderedDict()

# Load files created by 
for city in all:

    fname = 'analysis/anders/{0}/{1}'.format(scenario,city)
    
    file = open(fname, 'r')
    removed_ff[city],width[city],DX90[city],n[city] = pickle.load(file)
    file.close()
```
