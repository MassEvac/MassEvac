```python
import scipy.stats as ss
import cities
import massevac as me
import pickle

N = []
W = []

DX90 = []

T90ff = []
T90 = []

Qmeff = []
Qmebn = []

Qmdff = []
Qmdbn = []

Q90ff = []
Q90bn = []    

Q90 = []

Qpred = []

X = {}

def data(city):
    file = open('analysis/destins/{0}'.format(city), 'r')
    data = pickle.load(file)
    file.close()
    return data

def Q(city):
    file = open('analysis/destins_Q/{0}'.format(city), 'r')
    Q = pickle.load(file)
    file.close()
    return Q

def T(city):
    file = open('analysis/destins_T/{0}'.format(city), 'r')
    T = pickle.load(file)
    file.close()
    return T

def DX(city):
    file = open('analysis/destins_DX/{0}'.format(city), 'r')
    DX = pickle.load(file)
    file.close()
    return DX

def PW(city):
    file = open('analysis/width/{0}'.format(city), 'r')
    PW = pickle.load(file)
    file.close()
    return PW

this = 0
Xi = {}
```

```python
for city in cities.all:

    X[city], n, w, dx90, t90ff, t90, qmeff,qmebn, qmdff, qmdbn, q90ff, q90bn, q90, qpred = data(city)
    
    Xi[city] = range(this,this+len(X[city]))
    
    this = this+len(X[city])+1
    
    
    N.extend(n)        
    W.extend(w)

    DX90.extend(dx90)
    
    T90ff.extend(t90ff)    
    T90.extend(t90)
    
    Qmeff.extend(qmeff)
    Qmebn.extend(qmebn)
    
    Qmdff.extend(qmdff)    
    Qmdbn.extend(qmdbn)    
    
    Q90ff.extend(q90ff)
    Q90bn.extend(q90bn)
    
    Q90.extend(q90)
    
    Qpred.extend(qpred)
    
N_DX90 = np.array(N)/np.array(DX90)
```

```python
meanwidth = []

for city in cities.all:
    V = PW(city)
    for d in X[city]:
        try:
            meanwidth.append(mean(V[d][:1000]))
        except KeyError:
            meanwidth.append(0)
```

```python
# function that returns a vector of deviation from best line of fit
def deviation(x,y,p):
    y_mean = polyval(p,x)
    return abs(y_mean-y)
```

```python
# what order to use in the polyfit function?
prev=0
for order in range(1,10):
    p = polyfit(x,y,order)
    d = np.sqrt(sum(np.square(deviation(x,y,p))))
    print order,d,(prev/d-1)*100
    prev=d
    scatter(order,d)   
```

```python
# from the graph above, using higher order than 3 does not offer significant advantage in minimising deviation
p = polyfit(x,y,2)

plt.clf() # clear the figure
plt.scatter(x,y,c=deviation(x,y,p))
points = 100
x_range = range(points)/np.array(float(points))*max(x)
plt.plot(x_range,np.polyval(p,x_range),'r-',label='$y={0:0.3f}x^2+{1:0.3f}x+{2:0.3f}$'.format(*tuple(p))) # A red solid line
plt.legend(loc='lower center') # make a legend in the best location
plt.xlabel("Qpred") # labels again
plt.ylabel("Qmeff")

plt.xlabel("$Q_{p}(i)$") # labels again
plt.ylabel("$Q_{s}^{\overline{f}}(i)$")
plt.savefig('/home/bk12369/Dropbox/Thesis/2014.12.untitled/QsfVsQp.pdf')
```
