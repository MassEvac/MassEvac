# Evacuation time histogram

Here, we can see the histogram of evacuation time fitted with normal and log normal distribution curves. We can see that:

* In the free flow mode, normal distribution curve fits the histogram quite well, as well as the log normal distribution curve
* In the interactive flow mode, normal distribution is not a very good fit but the log normal distribution fits the histogram really well 
* Log normal distributions are compliant with a set of positive variables

```python
# This is the maximum time across all simulations
max_et = {}
for s in scenarios:
    max_et[s] = max(sim[s].T)
    
print max_et

# Establish number of bins proportional to the width of the page
count = {}
for s in scenarios:
    count[s] = int(100*max(sim[s].T)/max(max_et.values()))

print count    
```

# Flow per timestep

```python
# Ignore free flow case because it is rigged!
flow_vs_time= {}
flow_vs_density = {}
flow_hist = {}
F_tstep = {}

# Calculating flow from average density
for s in scenarios:
    F_tstep[s] = [D*common.f.velocity(D) for D in sim[s].average_D_tstep()]

    # Construct flow against time plot
    flow_vs_time[s] = plt.figure()
    plot(range(len(F_tstep[s])),F_tstep[s])
    plt.xlim(0,max(max_et.values()))
    xlabel('Time step')
    ylabel('Flow')
    plt.close(flow_vs_time[s])
    
    # Construct flow against density plot
    flow_vs_density[s] = plt.figure()
    scatter(sim[s].average_D_tstep(),F_tstep[s])
    xlabel('Density')
    ylabel('Flow')
    plt.close(flow_vs_density[s])
    
    # Construct flow histogram
    flow_hist[s] = plt.figure()
    hist(F_tstep[s],bins=5)
    xlabel('Average flow rate')
    ylabel('Number of time steps')
    plt.close(flow_hist[s])
```

```python
# Flow averaged afterwards
F = {}
for s in scenarios:
    F[s] = []
    for D_profile in sim[s].D_tstep:
        ratio = D_profile/float(sum(D_profile))
        F[s].append(sum([r*d*common.f.velocity(d) for r,d in zip(ratio,common.f.bin_mean)]))
```

```python
# Comparison of flow if calculated from average density or averaged afterwards
fig = plt.figure()
plt.scatter(F[ia],F_tstep[ia])
```
