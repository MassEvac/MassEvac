```python
import math
def lognorm_hist(label):
    print 'mean',np.mean(R[label])
    print 'std',np.std(R[label])    
    
    plt.figure()
    fontsize=15
    plt.hist(R[label],bins=50)
    plt.xlabel('$\mathrm{{{0} \ {1}}}$'.format(pl[label],unit[label]),fontsize=fontsize)
    plt.ylabel('$\mathrm{Number \ of \ CA}$',fontsize=fontsize)
    plt.savefig('../PhD/Papers/2014.12.untitled/figs/log-{0}.pdf'.format(label))        
    
    Nlog = np.log10(R[label])
    lmin = min(Nlog)-1
    lmax = max(Nlog)
        
    print '--'        
    print 'logmean',np.mean(Nlog)
    print 'logstd',np.std(Nlog)    
    print 'mean',10**np.mean(Nlog)
    print 'stdN1',10**np.mean(Nlog)-10**(np.mean(Nlog)-np.std(Nlog))
    print 'stdN2',10**(np.mean(Nlog)+np.std(Nlog))-10**np.mean(Nlog)
    
    plt.figure()
    bins=np.logspace(lmin, lmax, 50)
    plt.hist(R[label],bins=bins)
    plt.xlabel('$\mathrm{{{0} \ {1}}}$'.format(pl[label],unit[label]),fontsize=fontsize)
    plt.ylabel('$\mathrm{Number \ of \ CA}$',fontsize=fontsize)
    plt.xscale('log')
    plt.gcf().subplots_adjust(bottom=0.15)    
    plt.savefig('../PhD/Papers/2014.12.untitled/figs/log-{0}.pdf'.format(label))
```
