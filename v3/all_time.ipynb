```python
def plot_fig(l1,l2,c,norm=''):
    x = R[l1]
    y = R[l2]
    
    if norm:
        x = x/R[norm]
        y = y/R[norm]
        l1 += '.' + norm + '^{-1}'
        l2 += '.' + norm + '^{-1}'
    
    xl = np.linspace(min(x),max(x),10)   
    
    plt.figure()
    plt.plot(xl,xl,label='y=x')
    plt.scatter(x,y,label='R=%0.2f'%corrcoef(x,y)[0,1],c=c)
    plt.xlabel('${0}$'.format(l1))
    plt.ylabel('${0}$'.format(l2))
    plt.legend()
    corrcoef(x,y)
    plt.savefig('analysis/{0}/!figs/{1}-{2}.png'.format(sim,l1,l2))
```
